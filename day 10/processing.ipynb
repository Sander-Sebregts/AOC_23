# Init stuffs

### Imports

```python
from copy import deepcopy
```

### Input data

```python
filename = "./input_1.txt"

with open(file=filename) as f:
    input_data_pt1 = f.read().splitlines()

for index, line in enumerate(input_data_pt1):
    input_data_pt1[index] = [*line]
```

# Pt. 1

```python
# north = 0, east = 1, south = 2, west = 3
going = [
    ["|", "L", "J", "S"],
    ["-", "L", "F", "S"],
    ["|", "F", "7", "S"],
    ["-", "J", "7", "S"],
]
# connecting = going[(index+2)%4]
```

```python
simplified_maze = deepcopy(input_data_pt1)
modified = True

while modified:
    modified = False
    for y_coord, line in enumerate(simplified_maze):
        for x_coord, char in enumerate(line):
            if char == "." or char == "S":
                continue
            connecting_faces = [
                index for index, values in enumerate(going) if char in values
            ]

            # edge cases
            if (
                (x_coord == 0 and 3 in connecting_faces)
                or (x_coord == len(input_data_pt1[0]) - 1 and 1 in connecting_faces)
                or (y_coord == 0 and 0 in connecting_faces)
                or (y_coord == len(input_data_pt1) - 1 and 2 in connecting_faces)
            ):
                simplified_maze[y_coord][x_coord] = "."
                modified = True
            else:
                connecting = True
                if (
                    0 in connecting_faces
                    and simplified_maze[y_coord - 1][x_coord] not in going[2]
                ) or (
                    1 in connecting_faces
                    and simplified_maze[y_coord][x_coord+1] not in going[3]
                ) or (
                    2 in connecting_faces
                    and simplified_maze[y_coord + 1][x_coord] not in going[0]
                ) or (
                    3 in connecting_faces
                    and simplified_maze[y_coord][x_coord-1] not in going[1]
                ):
                    connecting = False

                if not connecting:
                    simplified_maze[y_coord][x_coord] = "."
                    modified = True
```

```python
# No complex redundant loops, remove the simple ones
for index, line in enumerate(simplified_maze):
    for char_index, char in enumerate(line):
        if (
            simplified_maze[index][char_index] == "F"
            and simplified_maze[index][char_index + 1] == "7"
            and simplified_maze[index + 1][char_index] == "L"
            and simplified_maze[index + 1][char_index + 1] == "J"
        ):
            simplified_maze[index][char_index] = "."
            simplified_maze[index][char_index + 1] = "."
            simplified_maze[index + 1][char_index] = "."
            simplified_maze[index + 1][char_index + 1] = "."

less_simplified_maze = deepcopy(simplified_maze)
for index, line in enumerate(less_simplified_maze):
    less_simplified_maze[index] = "".join(line)
less_simplified_maze
```

```python
for line_index, line in enumerate(input_data_pt1):
    if "S" in line:
        S_location = (line_index, line.index("S"))

DIRECTION_DICT = {
    0: {"|": 0, "F": 1, "7": 3},
    1: {"J": 0, "-": 1, "7": 2},
    2: {"|": 2, "J": 3, "L": 1},
    3: {"L": 0, "F": 2, "-": 3},
}
NEXT_Y = {0: -1, 1: 0, 2: 1, 3: 0}
NEXT_X = {0: 0, 1: 1, 2: 0, 3: -1}


def follow_maze_and_count_length(
    start_location: tuple[int, int], start_direction: int, maze: list[list]
) -> int:
    direction = start_direction
    location = start_location
    found_s_again = False
    loop_length = 0
    while not found_s_again:
        loop_length += 1
        location = (location[0] + NEXT_Y[direction], location[1] + NEXT_X[direction])
        next_char = maze[location[0]][location[1]]
        if next_char == "S":
            found_s_again = True
        else:
            direction = DIRECTION_DICT[direction][next_char]
    return loop_length


print(f"Half the loop length is: {follow_maze_and_count_length(start_location=S_location, start_direction=1, maze=simplified_maze)//2}")
```

# Pt. 2

```python
enclosed = 0
maze_copy = deepcopy(simplified_maze)

vertical_chars = ["|", "7", "J", "F", "L"]

def rules_on_verticals(verticals: list[str]) -> bool:
    current_length = len(verticals)
    subtract = 0
    for i in range(current_length):
        if (verticals[i] == "F" and verticals[i + 1] == "J") or (
            verticals[i] == "L" and verticals[i + 1] == "7"
        ):
            subtract += 1
    return (current_length - subtract) % 2 == 1


for y_index, line in enumerate(maze_copy):
    for x_index, curr_char in enumerate(line):
        if curr_char == ".":
            previous_verticals = [
                char for char in line[:x_index] if char in vertical_chars
            ]
            if rules_on_verticals(verticals=previous_verticals):
                maze_copy[y_index][x_index] = "I"
                enclosed += 1
            else:
                maze_copy[y_index][x_index] = "O"

print(f"Number of enclosed spaces: {enclosed}")
```

```python
for index, line in enumerate(maze_copy):
    maze_copy[index] = "".join(line)
maze_copy
```
